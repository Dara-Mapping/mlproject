```python
# f,ax=plt.subplots(1,2,figsize=(20,10))
# sns.countplot(x=df['race_ethnicity'],data=df,palette = 'bright',ax=ax[0],saturation=0.95)
# for container in ax[0].containers:
#     ax[0].bar_label(container,color='black',size=20)
    
# plt.pie(x = df['race_ethnicity'].value_counts(),labels=df['race_ethnicity'].value_counts().index,explode=[0.1,0,0,0,0],autopct='%1.1f%%',shadow=True)
# plt.show()   



# Set style
sns.set_style("whitegrid")

# Create figure with 1 row, 2 columns
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# ----- Countplot -----
sns.countplot(
    x='race_ethnicity',
    data=df,
    palette='bright',
    ax=ax[0],
    saturation=0.95
)

# Add labels on top of bars
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', fontsize=16)

ax[0].set_title("Race Ethnicity Count", fontsize=18)
ax[0].set_xlabel("Race Ethnicity", fontsize=14)
ax[0].set_ylabel("Count", fontsize=14)

# ----- Pie chart -----
race_counts = df['race_ethnicity'].value_counts()
colors = sns.color_palette('bright', len(race_counts))

# Optional: explode largest category
explode = [0.1 if i == race_counts.idxmax() else 0 for i in race_counts.index]

# Plot pie chart
wedges, texts, autotexts = ax[1].pie(
    race_counts,
    labels=race_counts.index,
    autopct='%1.1f%%',
    shadow=True,
    startangle=90,
    colors=colors,
    explode=explode,
    textprops={'fontsize': 16}
)
ax[1].set_title("Race/Ethnicity Distribution", fontsize=18)

# ----- Improve label contrast -----
for text, wedge in zip(autotexts, wedges):
    # If slice color is dark, use white font; otherwise black
    r, g, b = wedge.get_facecolor()[:3]  # RGB
    brightness = 0.299*r + 0.587*g + 0.114*b  # Perceived brightness
    text.set_color('white' if brightness < 0.6 else 'black')
    text.set_fontsize(16)
    
plt.tight_layout()
plt.show()
```
